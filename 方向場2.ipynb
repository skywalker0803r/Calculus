```python
import numpy as np
import matplotlib.pyplot as plt

# 定義常微分方程的右端項
def f(x, y):
    return -4*x

# 定義常微分方程的解
def exact_solution(x):
    return np.sqrt(-4*x**2 + 5)

# 定義繪製方向場的函數
def plot_direction_field(x_range, y_range):
    x = np.linspace(x_range[0], x_range[1], 20)
    y = np.linspace(y_range[0], y_range[1], 20)
    X, Y = np.meshgrid(x, y)
    U = 1
    V = f(X, Y)
    plt.quiver(X, Y, U, V, color='b', scale=20, label='Direction Field')

# 定義繪製解曲線的函數
def plot_exact_solution(x_range):
    x = np.linspace(x_range[0], x_range[1], 400)
    y = exact_solution(x)
    plt.plot(x, y, 'r', label='Exact Solution')

# 繪製方向場和解曲線
plt.figure(figsize=(8, 6))
plot_direction_field([-1.5, 1.5], [-3, 3])
plot_exact_solution([-1.5, 1.5])

# 繪製初始點
plt.scatter([1, 0], [1, 2], color='g', label='Initial Points (1, 1) and (0, 2)')

# 設置圖表屬性
plt.xlabel('x')
plt.ylabel('y')
plt.title('Direction Field and Exact Solution of the ODE')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
