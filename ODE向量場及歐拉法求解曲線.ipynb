```python
import numpy as np
import matplotlib.pyplot as plt

# 定義ODE的右手邊
def dy_dx(x, y):
    return -2*x*y

# 定義繪製向量場的函數
def plot_vector_field():
    x_range = np.linspace(-2, 2, 20)
    y_range = np.linspace(-2, 2, 20)
    X, Y = np.meshgrid(x_range, y_range)
    U = 1
    V = dy_dx(X, Y)
    # 正規化向量長度
    magnitude = np.sqrt(U**2 + V**2)
    U /= magnitude
    V /= magnitude
    plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=5, color='r')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Vector Field of the ODE')
    plt.grid(True)
    #plt.show()

# 繪製向量場
plot_vector_field()

# 定義Euler's method
def euler_method(dy_dx, x0, y0, h, num_points):
    x_values = [x0]
    y_values = [y0]
    for _ in range(num_points):
        y0 = y0 + h * dy_dx(x0, y0)
        x0 = x0 + h
        x_values.append(x0)
        y_values.append(y0)
    return np.array(x_values), np.array(y_values)

# 使用Euler's method計算解曲線，減小步長以提高計算精度
x0,y0 = 0,1.8
x_values, y_values = euler_method(dy_dx, x0, y0, 0.1, 20)

# 繪製解曲線
plt.plot(x_values, y_values, color='b', label='Solution Curve')
plt.scatter(x0,y0, color='g', label='Initial Point')
plt.legend()
plt.show()
```
