```python
import numpy as np
import matplotlib.pyplot as plt

# 定義ODE的右手邊
def dy_dx(x, y):
    return (y**2 - x**2) / (2 * x * y)

# 定義繪製向量場的函數
def plot_vector_field():
    x_range = np.linspace(-2, 2, 20)
    y_range = np.linspace(-2, 2, 20)
    X, Y = np.meshgrid(x_range, y_range)
    U = 1
    V = dy_dx(X, Y)
    # 正規化向量長度
    magnitude = np.sqrt(U**2 + V**2)
    U /= magnitude
    V /= magnitude
    plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=5, color='r')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Vector Field of the ODE')
    plt.grid(True)
    plt.show()

# 繪製向量場
plot_vector_field()
```
