```
import numpy as np
import matplotlib.pyplot as plt
from math import e

#1指數長的非常快
x = np.linspace(4,10,1000)

def factorial(n):
    if(n<=1):
        return 1
    else:
        return factorial(n-1) * n
y0 = x**x
y1 = [factorial(i) for i in x]
y2 = e**x
y3 = x**2
y4 = np.log(x)

for i,j in zip([y0,y1,y2,y3,y4],['y0','y1','y2','y3','y4']):
  plt.plot(i,label=j)
plt.legend()
plt.show()
```

```
for i,j in zip([y3,y4],['y3','y4']):
  plt.plot(i,label=j)
plt.legend()
plt.show()
```
