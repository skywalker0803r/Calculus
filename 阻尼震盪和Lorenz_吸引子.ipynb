```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# 定義二階常微分方程 - 阻尼擺動
def damping_pendulum(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

# 定義初始條件和時間範圍
y0 = [np.pi - 0.1, 0.0]  # 初始角度和角速度
t = np.linspace(0, 50, 1000)  # 模擬的時間範圍

# 計算阻尼擺動的解
b = 0.25  # 阻尼係數
c = 5.0   # 摩擦係數
sol = odeint(damping_pendulum, y0, t, args=(b, c))

# 繪製解曲線
plt.figure(figsize=(8, 6))
plt.plot(t, sol[:, 0], 'b', label='Angle(t)')
plt.plot(t, sol[:, 1], 'g', label='Angular Velocity(t)')
plt.legend(loc='best')
plt.xlabel('Time')
plt.title('Damped Pendulum Motion')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import odeint

# 定義阻尼擺動的常微分方程
def damping_pendulum(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

# 初始條件和時間範圍
y0 = [np.pi - 0.1, 0.0]  # 初始角度和角速度
t = np.linspace(0, 50, 1000)  # 模擬的時間範圍

# 計算阻尼擺動的解
b = 0.25  # 阻尼係數
c = 5.0   # 摩擦係數
sol = odeint(damping_pendulum, y0, t, args=(b, c))

# 繪製解曲線
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:, 0], sol[:, 1], t, label='Damped Pendulum Solution')
equation_text = r'$\frac{{d^2\theta}}{{dt^2}} = -b\frac{{d\theta}}{{dt}} - c\sin(\theta)$'
ax.text(0, 2, 30, equation_text, fontsize=12, color='black')
ax.set_xlabel('Theta')
ax.set_ylabel('Omega')
ax.set_zlabel('Time')
ax.set_title('Damped Pendulum Motion in 3D')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# 定義 Lorenz 吸引子的二階常微分方程
def lorenz_system(y, t, sigma, rho, beta):
    y1, y2, y3 = y
    dydt = [sigma * (y2 - y1), y1 * (rho - y3) - y2, y1 * y2 - beta * y3]
    return dydt

# 定義初始條件和參數
y0 = [1, 0, 0]
sigma, rho, beta = 10, 28, 8/3

# 定義時間範圍
t = np.linspace(0, 50, 10000)

# 計算 Lorenz 吸引子的解
sol = odeint(lorenz_system, y0, t, args=(sigma, rho, beta))

# 繪製 Lorenz 吸引子的三維軌跡
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')
ax.plot(sol[:, 0], sol[:, 1], sol[:, 2], lw=0.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor")
# 添加數學表達式標籤
equation_text = r'$\frac{dx}{dt} = \sigma(y - x)$''\n' r'$\frac{dy}{dt} = x(\rho - z) - y$''\n' r'$\frac{dz}{dt} = xy - \beta z$'
ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# 定義 Lorenz 吸引子的二階常微分方程
def lorenz_system(y, t, sigma, rho, beta):
    y1, y2, y3 = y
    dydt = [sigma * (y2 - y1), y1 * (rho - y3) - y2, y1 * y2 - beta * y3]
    return dydt

# 定義初始條件和參數
y0 = [1, 0, 0]
sigma, rho, beta = 10, 28, 8/3

# 定義時間範圍
t = np.linspace(0, 50, 10000)

# 計算 Lorenz 吸引子的解
sol = odeint(lorenz_system, y0, t, args=(sigma, rho, beta))

# 繪製 Lorenz 吸引子的三維軌跡，為解曲線指定多個顏色
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 定義多個顏色
colors = plt.cm.viridis(np.linspace(0, 1, len(sol)))

# 繪製解曲線，每段曲線使用不同的顏色
for i in range(len(sol) - 1):
    ax.plot(sol[i:i + 2, 0], sol[i:i + 2, 1], sol[i:i + 2, 2], color=colors[i], lw=0.5)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Lorenz Attractor with Multiple Colors")
# 添加數學表達式標籤
equation_text = r'$\frac{dx}{dt} = \sigma(y - x)$''\n' r'$\frac{dy}{dt} = x(\rho - z) - y$''\n' r'$\frac{dz}{dt} = xy - \beta z$'
ax.text2D(0.05, 0.95, equation_text, transform=ax.transAxes, fontsize=12, verticalalignment='top')

plt.show()
```
