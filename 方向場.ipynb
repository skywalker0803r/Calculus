```python
import numpy as np
import matplotlib.pyplot as plt

# 
def dy_dx(x, y):
    return 1 + y**2

# Euler's method
def euler_method(dy_dx, x0, y0, h, num_points):
    x_values = [x0]
    y_values = [y0]
    for _ in range(num_points):
        y0 = y0 + h * dy_dx(x0, y0)
        x0 = x0 + h
        x_values.append(x0)
        y_values.append(y0)
    return np.array(x_values), np.array(y_values)

# 
x = np.linspace(0, 2, 20)
y = np.linspace(-1, 2, 20)
X, Y = np.meshgrid(x, y)
U = 1
V = 1 + Y**2

# 
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color='b', scale=20)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Direction Field')
plt.grid(True)

# Euler's method
x_values, y_values = euler_method(dy_dx, np.pi/4, 1, 0.01, 30)

# 
plt.plot(x_values, y_values, color='r', label='Solution Curve')
plt.scatter([np.pi/4], [1], color='g', label='Initial Point')
plt.legend()
plt.show()
```
