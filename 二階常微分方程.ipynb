```python
import numpy as np
import matplotlib.pyplot as plt

# 
def second_order_diff_eq(y, t):
    #  y'' = -y
    y1, y2 = y
    return [y2, -y1]

# 
def vector_field(y, t):
    y1, y2 = y
    return [y2, -y1]

# 
t = np.linspace(0, 10, 1000)

# 
from scipy.integrate import odeint
y0 = [1, 0]  #  y(0) = 1, y'(0) = 0
sol = odeint(second_order_diff_eq, y0, t)

# 
y1, y2 = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
u, v = np.zeros_like(y1), np.zeros_like(y2)
for i in range(len(y1)):
    for j in range(len(y2)):
        x_dot, y_dot = vector_field([y1[i, j], y2[i, j]], 0)
        u[i,j] = x_dot
        v[i,j] = y_dot

# 
plt.figure(figsize=(8, 6))
plt.plot(sol[:, 0], sol[:, 1], label='Solution Curve')

# 
plt.quiver(y1, y2, u, v, color='r', scale=20, scale_units='inches')

# 
plt.xlabel('y')
plt.ylabel("y'")
plt.title('Second Order Differential Equation Solution Curve and Vector Field')
plt.legend()
plt.grid(True)
plt.show()
```
