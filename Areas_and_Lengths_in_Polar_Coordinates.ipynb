```python
import matplotlib.pyplot as plt
from math import sin,cos,radians
import numpy as np

def f(r,θ):
  θ = radians(θ)
  x = r*cos(θ)
  y = r*sin(θ)
  return x,y
plt.style.use('bmh')
```

# example1

```python
def example1a(θ):
  θ = radians(θ)
  r = 3*sin(θ)
  return r

def example1b(θ):
  θ = radians(θ)
  r = 1+sin(θ)
  return r

θ = np.linspace(0,360,1000)
r1 = [example1a(i) for i in θ]
r2 = [example1b(i) for i in θ]
x1 = [f(i,j)[0] for i,j in zip(r1,θ)]
y1 = [f(i,j)[1] for i,j in zip(r1,θ)]
x2 = [f(i,j)[0] for i,j in zip(r2,θ)]
y2 = [f(i,j)[1] for i,j in zip(r2,θ)]
plt.xlim(np.min(x1+x2)-1,np.max(x1+x2)+1)
plt.ylim(np.min(y1+y2)-1,np.max(y1+y2)+1)
plt.scatter(x1,y1,c='r', s=5, alpha=0.8)
plt.scatter(x2,y2,c='b', s=5, alpha=0.8)
plt.legend()
plt.fill_between(x1, y1,color='gray')
plt.fill_between(x2, y2,color='w')
plt.show()
```

```python
from __future__ import division
from sympy import *
from math import pi
```

```python
def obj(x):
  return (1/2) * ((3*sin(x))**2 - (1+sin(x))**2)
x = symbols('x', communtative = True)
```

```python
print(integrate(obj(x), (x, pi/6, 5*pi/6)))
print(pi)
```

```python
def example2a(θ):
  θ = radians(θ)
  r = sin(2*θ)
  return r

def example2b(θ):
  θ = radians(θ)
  r = cos(2*θ)
  return r

θ = np.linspace(0,360,1000)
r1 = [example2a(i) for i in θ]
r2 = [example2b(i) for i in θ]
x1 = [f(i,j)[0] for i,j in zip(r1,θ)]
y1 = [f(i,j)[1] for i,j in zip(r1,θ)]
x2 = [f(i,j)[0] for i,j in zip(r2,θ)]
y2 = [f(i,j)[1] for i,j in zip(r2,θ)]
plt.scatter(x1,y1,c='r', s=1, alpha=0.8)
plt.scatter(x2,y2,c='b', s=1, alpha=0.8)
plt.legend()
plt.show()
```
