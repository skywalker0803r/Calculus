```python
import matplotlib.pyplot as plt
from math import sin,cos,radians
import numpy as np

def f(r,θ):
  θ = radians(θ)
  x = r*cos(θ)
  y = r*sin(θ)
  return x,y
plt.style.use('bmh')

x,y = f(1,30)

plt.xlim([0,1])
plt.ylim([0,1])
plt.scatter(0,0,c='b')
plt.scatter(x,y,c='b')
plt.plot([0,x],[0,y],c='r')
plt.show()
```

```python
x,y = f(-1,180+30)

plt.xlim([0,1])
plt.ylim([0,1])
plt.scatter(0,0,c='b')
plt.scatter(x,y,c='b')
plt.plot([0,x],[0,y],c='r')
plt.show()
```

# example 3

```python
def example3(θ):
  θ = radians(θ)
  r = 2*cos(4*θ)
  return r

θ = np.linspace(0,360,1000)
r = [example3(i) for i in θ]
x = [f(i,j)[0] for i,j in zip(r,θ)]
y = [f(i,j)[1] for i,j in zip(r,θ)]
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.scatter(x,y,c=np.arctan2(x,y), cmap='rainbow', s=50, alpha=0.8)
plt.legend()
plt.show()
```

# example4

```python
def example4(θ):
  θ = radians(θ)
  r = 1 + sin(θ)
  return r

θ = np.linspace(0,360,1000)
r = [example4(i) for i in θ]
x = [f(i,j)[0] for i,j in zip(r,θ)]
y = [f(i,j)[1] for i,j in zip(r,θ)]
plt.xlim([np.min(x)*1.5,np.max(x)*1.5])
plt.ylim([np.min(y)*1.5,np.max(y)*1.5])
plt.scatter(x,y,c=np.arctan2(x,y), cmap='rainbow', s=50, alpha=0.8)
plt.legend()
plt.show()
```
